# basemap

*basemap* is built as an extension of  [matplotlib](http://matplotlib.org/). It superimposes a geographical coordinate system on the top of matplotlib's one. It is usually used with modules such as [pyshp](https://pypi.python.org/pypi/pyshp) oto handle [shapefiles](https://en.wikipedia.org/wiki/Shapefile).

[documentation](http://matplotlib.org/basemap/) [source](https://github.com/matplotlib/basemap) [installation](http://matplotlib.org/basemap/users/installing.html) [tutorial](https://basemaptutorial.readthedocs.io/en/latest/) [gallery](http://matplotlib.org/basemap/users/examples.html)

## example

```python
from mpl_toolkits.basemap import Basemap
import numpy
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 1, figsize=(8,8))
m = Basemap(llcrnrlon=-5,llcrnrlat=40,urcrnrlon=20,urcrnrlat=56,
            resolution='i',projection='cass',lon_0=2.34,lat_0=48,
           ax=axes)
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='lightgrey', lake_color='#AAAAFF')
m.drawparallels(numpy.arange(-40,61.,2.))
m.drawmeridians(numpy.arange(-20.,21.,2.))
m.drawmapboundary(fill_color='#BBBBFF')
```
