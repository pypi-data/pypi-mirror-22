```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
# Gravity Recovery and Climate Experiment (GRACE) Data
# Source: http://grace.jpl.nasa.gov/
# Current surface mass change data, measuring equivalent water thickness in cm, versus time
# with a resolution of 1x1 degree
```

```python
from skdaccess.geo.grace import DataFetcher as GR_DF
from skdaccess.framework.param_class import *
```

```python
geo_point = AutoParam((38, -117)) # location in Nevada
grace_fetcher = GR_DF([geo_point],start_date='2010-01-01',end_date='2014-01-01',resample=False)
```

```python
grace_data = grace_fetcher.output().get() # returns a pandas data panel
```

```python
grace_data.head()
```

```python
plt.figure();
plt.plot(grace_data['Grace']);
plt.tight_layout()
```
